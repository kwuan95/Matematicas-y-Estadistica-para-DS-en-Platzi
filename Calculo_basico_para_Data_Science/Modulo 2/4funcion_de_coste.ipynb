```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = np.array([
  [0, 100],
  [20, 150],
  [30, 200],
  [40, 180],
  [50, 250],
  [60, 230]
])

df = pd.DataFrame(df)
plt.scatter(df[0], df[1])
plt.show()
```

```python
x = df[0]
y = df[1]

m, b = np.polyfit(x, y, 1)

plt.scatter(x, y)
plt.plot(x, m*x + b)
plt.show()
```

```python
y_pred = m*x + b

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("MSE:", mse)
print("R2:", r2)
```
