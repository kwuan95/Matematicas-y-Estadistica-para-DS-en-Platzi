## Propiedades de la norma



```python
v1 = np.array([2,7])
v2 = np.array([3,5])

print(v1)
print(v2)
```

```python
v1v2 = v1+v2
print(v1v2)
```

```python
np.linalg.norm(v1v2)
```

```python
# norma(v1v2) <= norma(v1) + norma(v2)
normav1 = np.linalg.norm(v1)
normav2 = np.linalg.norm(v2)
normav1v2 = np.linalg.norm(v1v2)
print(normav1v2 <= normav1 + normav2)
```

```python
print(normav1+normav2)
```

```python
v1 = np.array([0,0,2,7])
v2 = np.array([0,0,3,5])

v1aux = np.array([v1[2],v1[3],v2[2],v2[3]])

v1v2 = np.array([0,0,5,12])
```

```python
plt.quiver([v1[0], v1aux[0], v1v2[0]],
           [v1[1], v1aux[1], v1v2[1]],
           [v1[2], v1aux[2], v1v2[2]],
           [v1[3], v1aux[3], v1v2[3]],
           angles = 'xy', scale_units ='xy', scale=1, color=sns.color_palette())

plt.xlim(-0.5,6)
plt.ylim(-0.5,15)
```
