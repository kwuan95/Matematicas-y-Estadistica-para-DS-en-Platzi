```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%run "./graficarVectores.ipynb"
```

```python
v1 = np.array([0,0,0,3])
v2 = np.array([0,0,3,3])
```

```python
plt.xlim(-2,6)
plt.ylim(-2,6)

plt.quiver( [v1[0], v2[0]],
            [v1[1], v2[1]],
            [v1[2], v2[2]],
            [v1[3], v2[3]],
    angles = 'xy', scale_units ='xy', scale=1, color=sns.color_palette())
```

```python
v1 = np.array([0,3])
v2 = np.array([3,3])

print(v1.T.dot(v2))
```

```python
normav1 = np.linalg.norm(v1)
normav2 = np.linalg.norm(v2)

print(normav1 * normav2 * np.cos(np.deg2rad(45)))
```
