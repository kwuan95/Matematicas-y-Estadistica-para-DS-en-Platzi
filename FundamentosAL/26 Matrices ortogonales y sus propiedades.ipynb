```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%run "./graficarVectores.ipynb"
```

```python
matriz = np.array([[1,0,0],
                   [0,1,0],
                   [0,0,1]])
print(matriz)
```

### Validar si todas los vectores de la matriz son ortogonales

```python
print(matriz[:,0].dot(matriz[:,1]))
```

```python
print(matriz[:,0].dot(matriz[:,2]))
```

```python
print(matriz[:,1].dot(matriz[:,2]))
```

### Confirmacion de ue todos los vectores de la matriz son ortonomales (la norma da 1)

```python
print(np.linalg.norm(matriz[:,0]))
print(np.linalg.norm(matriz[:,1]))
print(np.linalg.norm(matriz[:,2]))
```

### Los mismos calculos pero para las filas

```python
print(matriz[0,:].dot(matriz[1,:]))
print(matriz[0,:].dot(matriz[2,:]))
print(matriz[1,:].dot(matriz[2,:]))
```

```python
print(np.linalg.norm(matriz[0,:]))
print(np.linalg.norm(matriz[1,:]))
print(np.linalg.norm(matriz[2,:]))
```

```python
A = np.array([[np.cos(100), -np.sin(100)],
              [np.sin(100), np.cos(100)]])
A
```

```python
print(np.linalg.norm(A[0,:]))
print(np.linalg.norm(A[1,:]))
print(np.linalg.norm(A[:,0]))
print(np.linalg.norm(A[:,1]))
```

```python
print(A[0,:].dot(A[1,:]))
print(A[:,0].dot(A[:,1]))
```

```python
At = A.T
print(At.dot(A))
print(A.dot(At))
```

```python
Ainv = np.linalg.inv(A)
print(Ainv)
```
