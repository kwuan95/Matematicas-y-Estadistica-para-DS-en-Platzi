```python
import numpy as np
```

```python
# esto hace q si algun calculo me da un numero muy cercano a 0, q ignore los decimales y me de un 0
np.set_printoptions(suppress=True)
```

```python
A = np.array([[3,1],[2,1]])
A
```

```python
b = np.array([[1],[1]])
```

```python
inversa_A = np.linalg.inv(A)
inversa_A
```

```python
x = inversa_A.dot(b)
x
```

```python
print(A.dot(x))
```

```python
sol_2 = inversa_A.dot(np.array([[3],[7]]))
sol_2
```

```python
print(A.dot(sol_2))
```
