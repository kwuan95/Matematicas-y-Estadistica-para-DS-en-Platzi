```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sin Solucion

```python
x = np.arange(-6,6)

y1 = 3*x + 5
y2 = -x + 3
y3 = 2*x + 1
```

```python
plt.figure()

plt.plot(x,y1,color='green')
plt.plot(x,y2,color='blue')
plt.plot(x,y3,color='red')

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axhline(y=0, color='grey')
plt.axvline(x=0, color='grey')
plt.show()

```

no tiene solucion ya q no existe punto en el cual se cruzen las 3 rectas (Sobredeterminado)

## Una Solucion

```python
x = np.arange(-6,6)

y2 = -x + 3
y3 = 2*x + 1
```

```python
plt.figure()

plt.plot(x,y2,color='blue')
plt.plot(x,y3,color='red')

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axhline(y=0, color='grey')
plt.axvline(x=0, color='grey')
plt.show()
```

un cruze entre rectas, es la solucion (determinado)

## Infinitas soluciones

```python
plt.figure()

plt.plot(x,y3,color='red')

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axhline(y=0, color='grey')
plt.axvline(x=0, color='grey')
plt.show()
```

todos los valores son una solucion valida (indeterminado)
